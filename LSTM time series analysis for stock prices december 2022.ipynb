```python
df['Date']= pd.to_datetime(df['Date'])
```

```python
filtered_df = df.loc[(df['Date'] >= '2017-01-01')
                     & (df['Date'] < '2022-12-31')]
```

## We will predict here the opening price

```python
training_set=filtered_df.iloc[:1409,1:2].values
```

### We need to scale our data as LSTM need scaled data

```python
training_set
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
training_set = scaler.fit_transform(training_set)
```

```python
training_set
```

```python
x_train = training_set[0:1408]
y_train = training_set[1:1409]
```

```python
x_train = np.reshape(x_train,(1408,1,1))
```

```python
x_train
```

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense
```

```python
model = Sequential()
model.add(LSTM(units=4,activation='sigmoid',input_shape=(None,1)))
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(x_train,y_train,batch_size=64,epochs=200)
```

```python
df
```

```python
filtered_df
```

```python

test_set = filtered_df.iloc[1409:,1:2].values
```

```python
test_set
```

```python
inputs=test_set
```

```python
inputs=scaler.transform(inputs)
```

```python
inputs=np.reshape(inputs,(100,1,1))
```

```python
predicted_stock_price = model.predict(inputs)
```

```python
predicted_stock_price=scaler.inverse_transform(predicted_stock_price)
```

```python
predicted_stock_price
```

```python
plt.plot(test_set,color='red',label='Real Stock prices for Apple 2023')
plt.plot(predicted_stock_price,color='blue',label='Predicted Stock prices for Apple 2023')
plt.title('Apple Stock price prediction')
plt.xlabel('Time')
plt.ylabel('Apple Stock Price')
plt.legend()
plt.show()
```

### We see here that we didn't predict the stock prices but actually we predicted the trends in the stock prices for the last 30 days in the year
